### 2. PREPARACIÓN DATASET COUNTRIES

```python
df_count = pd.read_csv("../data/countries of the world.csv")

df_count.info()

df_count.head(10)
```

#### 2.1 Limpieza de formato de datos y columnas
- [ ] Eliminar espacios en blanco en nombres de columnas. Estandarizar nombres a formato snake_case.
- [ ] Comprobar si los separadores decimales están en formato correcto (algunas versiones usan coma en vez de punto).
- [ ] Conversión de tipos
- [ ] Convertir a `category`:
  - Country
  - Region
  - Climate (variable categórica codificada)

```python
# Normalización de los nombres de las columnas
df_count = df_count.rename(columns={'Area (sq. mi.)': 'Area_sq_mi',
    'Pop. Density (per sq. mi.)': 'Pop_density',
    'Coastline (coast/area ratio)': 'Coastline',
    'Net migration': 'Migration',
    'Infant mortality (per 1000 births)': 'Infant_mortality',
    'GDP ($ per capita)': 'GDP',
    'Literacy (%)': 'Literacy',
    'Phones (per 1000)': 'Phones',
    'Arable (%)': 'Arable',
    'Crops (%)': 'Crops',
    'Other (%)': 'Other'
    }
)
```

```python
def search_cols_with_commas(df):
    cols = []
    for col in df.columns:
        if df[col].dtype == 'object' and df[col].str.contains(',', regex=False, na=False).any():
            cols.append(col)
    return cols

cols_with_commas = search_cols_with_commas(df_count)
print(cols_with_commas)
```

```python
del cols_with_commas[0] # Eliminamos la columan Country porque realmente es string
print(cols_with_commas)
```

```python
# Convertir los valores numéricos de string a float
for col in cols_with_commas:
    df_count[col] = (
        df_count[col]
        .str.replace(',', '.', regex=False)
        .astype(float)
    )
```

Para optimizar operaciones como groupby, value_counts. 
Evitar errores al no permitir texto libre. 
Especificar que no tienen carácter numérico
- [ ] Convertir a `category` :
  - Country
  - Region
  - Climate (variable categórica codificada)

```python
df_count["Country"] = df_count["Country"].astype("category")
df_count["Region"] = df_count["Region"].astype("category")
# para categorizar la columna clima nos quedaremos con el primer valor ya que no es un factor con suficiente peso en nuestro EDA
df_count["Climate"] = (
    df_count["Climate"]
    .astype(str)
    .str.split(",", n=1)
    .str[0]
    .str.strip()
    .replace({"nan": None})
    .astype("category")
)
df_count.info()
```

#### 2.2 Limpieza de formato de datos y columnas
- [ ] Identificar columnas con `NaN`.
- [ ] Decidir estrategia:
  - Imputación (media/mediana) para variables numéricas.
  - Imputación por moda para categóricas.
  - Eliminación de filas si el porcentaje de nulos es bajo.
- [ ] Documentar la decisión (importante para reproducibilidad).

```python
# Identificar columans con NaN
cols_con_nulos = df_count.columns[df_count.isna().any()]
cols_con_nulos
```

```python
def replace_nan_by_mean(df, columns, group_col="Region"):
    for col in columns:
        # Sólo reemplazar si la columna es numérica
        if df[col].dtype.kind in "fi":  # float o int
            df[col] = df[col].fillna(
                df.groupby(group_col)[col].transform("mean")
            )
    return df
```

```python
df_cof_treated = replace_nan_by_mean(df_count, cols_con_nulos)

df_cof_treated.info()
```

#### 3. Merge de los dos datasets

- Limpiar y dividir la columna Countries en el dataset **df_lang** en listas
- Explode: cada país de la lista → su propia fila
- Limpiar espacios en la nueva columna y normalizar capitalización
- Limpiar y dividir la columna Countries en el dataset **df_cow** en listas
- Mapeo para que coincidan los nombres de los países en los dos datasets
- Merge
- Diagnóstico: cuántas lenguas quedaron sin datos de país

```python
# 1 Limpiar y dividir la columna Countries en el dataset **df_lang** en listas

df_lang['Countries'] = df_lang['Countries'].str.strip()
df_lang['Countries_list'] = df_lang['Countries'].str.split(',')


# 2 Explode: cada país de la lista → su propia fila

df_lang_exploded = df_lang.explode('Countries_list')


# 3 Limpiar espacios en la nueva columna y normalizar capitalización
df_lang_exploded['Countries_list'] = df_lang_exploded['Countries_list'].str.strip().str.title()


# 4 Limpiar y dividir la columna Countries en el dataset **df_cow** en listas
df_cof_treated['Country'] = df_cof_treated['Country'].str.strip().str.title()

# 5. Mapeo para que coincidan los nombres de los países en los dos datasets
mapeo = {
    "United States Of America": "United States",
    "Russian Federation": "Russia",
    "Bolivia (Plurinational State Of)": "Bolivia",
    "Venezuela (Bolivarian Republic Of)": "Venezuela",
    "Lao People'S Democratic Republic": "Laos",
    "Myanmar": "Burma",
    "Viet Nam": "Vietnam",
    "Iran (Islamic Republic Of)": "Iran",
    "New Caledonia (France)": "New Caledonia",
    "Micronesia (Federated States Of)": "Micronesia, Fed. St.",
    "United Kingdom Of Great Britain And Northern Ireland": "United Kingdom",
    "The Former Yugoslav Republic Of Macedonia": "Macedonia",
    "United Republic Of Tanzania": "Tanzania",
    "Syrian Arab Republic": "Syria",
    "Libyan Arab Jamahiriya": "Libya",
    "Democratic Republic Of The Congo": "Congo, Dem. Rep.",
    "Timor-Leste": "East Timor",
    "Central African Republic": "Central African Rep.",
    "French Polynesia (France)": "French Polynesia",
    "French Guiana (France)": "French Guiana",
    "Côte D'Ivoire": "Ivory Coast",
    "Greenland (Kingdom Of Denmark)": "Greenland",
    "Bosnia And Herzegovina": "Bosnia & Herzegovina",
    "Republic Of Moldova": "Moldova",
    "Congo": "Congo, Repub. Of The",
    "Republic Of Korea": "Korea, South",
    "Guam (U.S.A.)": "Guam",
}

# 6. Aplicar mapeo antes del merge
df_lang_exploded['Countries_list'] = df_lang_exploded['Countries_list'].replace(mapeo)


# 7. Merge
df_merged = df_lang_exploded.merge(
    df_cof_treated,
    left_on='Countries_list',
    right_on='Country',
    how='left'
)

# 8. Diagnóstico: cuántas lenguas quedaron sin datos de país
sin_pais = df_merged['Country'].isna().sum()
print(f"Filas sin match de país: {sin_pais} ({sin_pais/len(df_merged)*100:.1f}%)")
print(f"Filas totales tras explode + merge: {len(df_merged)}")
```
