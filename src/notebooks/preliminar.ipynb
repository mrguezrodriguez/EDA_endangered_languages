```python
import pandas as pd

# 1. Cargar los datos
df_lang = pd.read_csv("../data/languages.csv")
df_countries = pd.read_csv("../data/countries of the world.csv")

# 2. Limpiar y dividir la columna Countries en listas
df_lang['Countries'] = df_lang['Countries'].str.strip()
df_lang['Countries_list'] = df_lang['Countries'].str.split(',')

# 3. Explode: cada país de la lista → su propia fila
df_lang_exploded = df_lang.explode('Countries_list')

# 4. Limpiar espacios en la nueva columna y normalizar capitalización
df_lang_exploded['Countries_list'] = df_lang_exploded['Countries_list'].str.strip().str.title()

# 5. Lo mismo para el dataset de países
df_countries['Country'] = df_countries['Country'].str.strip().str.title()

# 6. Mapeo para que coincidan los nombres de los países en los dos datasets
mapeo = {
    "United States Of America": "United States",
    "Russian Federation": "Russia",
    "Bolivia (Plurinational State Of)": "Bolivia",
    "Venezuela (Bolivarian Republic Of)": "Venezuela",
    "Lao People'S Democratic Republic": "Laos",
    "Myanmar": "Burma",
    "Viet Nam": "Vietnam",
    "Iran (Islamic Republic Of)": "Iran",
    "New Caledonia (France)": "New Caledonia",
    "Micronesia (Federated States Of)": "Micronesia, Fed. St.",
    "United Kingdom Of Great Britain And Northern Ireland": "United Kingdom",
    "The Former Yugoslav Republic Of Macedonia": "Macedonia",
    "United Republic Of Tanzania": "Tanzania",
    "Syrian Arab Republic": "Syria",
    "Libyan Arab Jamahiriya": "Libya",
    "Democratic Republic Of The Congo": "Congo, Dem. Rep.",
    "Timor-Leste": "East Timor",
    "Central African Republic": "Central African Rep.",
    "French Polynesia (France)": "French Polynesia",
    "French Guiana (France)": "French Guiana",
    "Côte D'Ivoire": "Ivory Coast",
    "Greenland (Kingdom Of Denmark)": "Greenland",
    "Bosnia And Herzegovina": "Bosnia & Herzegovina",
    "Republic Of Moldova": "Moldova",
    "Congo": "Congo, Repub. Of The",
    "Republic Of Korea": "Korea, South",
    "Guam (U.S.A.)": "Guam",
}

# Aplicar mapeo antes del merge
df_lang_exploded['Countries_list'] = df_lang_exploded['Countries_list'].replace(mapeo)


# 7. Merge
df_merged = df_lang_exploded.merge(
    df_countries,
    left_on='Countries_list',
    right_on='Country',
    how='left'
)

# 8. Diagnóstico: cuántas lenguas quedaron sin datos de país
sin_pais = df_merged['Country'].isna().sum()
print(f"Filas sin match de país: {sin_pais} ({sin_pais/len(df_merged)*100:.1f}%)")
print(f"Filas totales tras explode + merge: {len(df_merged)}")
```
